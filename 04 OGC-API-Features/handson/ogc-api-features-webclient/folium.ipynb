# OGC API Features in Jupyter Notebook / Folium

Folium is een python library om kaartjes in een jupyter notebook te laten zien.

Om deze demo te uit te voeren heb je een python omgeving nodig met:
- jupyter notebook
- folium
- geopandas
- requests

__Het eerste voorbeeld gebruikt de buurten uit de NodeJS OGC API Features server die in bouw_je_eigen_server staat__


```python
import folium
import requests  # For making HTTP requests
import geopandas as gpd
import json
```

```python
m = folium.Map(location=[52.3, 4.8], zoom_start=12)

url = "http://localhost:8080/demoservice/v1/collections/buurten/items?f=json&limit=1000"

# Fetch the GeoJSON data
gdf = gpd.read_file(url)

# Convert to GeoJSON
geojson = gdf.to_json()

folium.GeoJson(
    gdf,
    name="buurten",
    style_function=lambda feature: {
        # "fillColor": "#ffff00",
        "color": "blue",
        "weight": 2,
        "dashArray": "5, 5",
    },
).add_to(m)

m
```

## nog een voorbeeld

```python


url = "https://groningerdorpen.nl/wp-json/ogc/v1/lp/collections/dorpshuizen/items"

# Fetch the GeoJSON data
gdf = gpd.read_file(url)

gdf.head()
```

```python
m = folium.Map(location=[53.2, 6.5], zoom_start=10)

#quick hack omdat de last_update kolom niet goed naar geojson geconverteerd wordt...
gdf_subset = gdf[['id','title','dorp','gemeente', 'geometry']]

# Convert to GeoJSON
geojson = gdf_subset.to_json()

folium.GeoJson(
    gdf_subset,
    name="dorpshuizen",
    style_function=lambda x: {'fillColor': 'blue', 'color': 'black', 'weight': 1, 'fillOpacity': 0.5},
    tooltip=folium.GeoJsonTooltip(fields=["id"]),
    popup=folium.GeoJsonPopup(fields=["id", "title"]),
    highlight_function=lambda x: {"fillOpacity": 0.8},
).add_to(m)

m


```
